```python
#Ensure sales_in_thousands in at least 25k 
for x in df.index: 
         if df.loc[x,'Sales_in_thousands'] < 25:
            df.loc[x,'Sales_in_thousands'] == 25
print(df[('Sales_in_thousands')])
```

```python
#Removing row with wrong data 
for x in df.index: 
         if df.loc[x, 'Sales_in_thousands'] < 25:
            df.drop(x) 
print(df[('Sales_in_thousands')])
```
